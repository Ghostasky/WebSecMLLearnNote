```python
import sys
from urllib.parse import urlparse
import re
# from hmmlearn import hmm
import numpy as np
# from sklearn.externals import joblib
# import HTMLParser
import nltk
import csv
import matplotlib.pyplot as plt
from nltk.probability import FreqDist
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
```

```python
#测试样本数
N=90

def load_user_cmd_new(filename):
    cmd_list=[]
    dist=[]
    with open(filename) as f:
        i=0
        x=[]
        for line in f:
            line=line.strip('\n')
            x.append(line)
            dist.append(line)
            i+=1
            if i == 100:
                cmd_list.append(x)
                x=[]
                i=0

    fdist = FreqDist(dist).keys()
    return cmd_list,fdist

def load_user_cmd(filename):
    cmd_list=[]
    dist_max=[]
    dist_min=[]
    dist=[]
    with open(filename) as f:
        i=0
        x=[]
        for line in f:
            line=line.strip('\n')
            x.append(line)
            dist.append(line)
            i+=1
            if i == 100:
                cmd_list.append(x)
                x=[]
                i=0

    fdist = FreqDist(dist).keys()
    dist_max=set(fdist[0:50])
    dist_min = set(fdist[-50:])
    return cmd_list,dist_max,dist_min

def get_user_cmd_feature(user_cmd_list,dist_max,dist_min):
    user_cmd_feature=[]
    for cmd_block in user_cmd_list:
        f1=len(set(cmd_block))
        fdist = FreqDist(cmd_block).keys()
        f2=fdist[0:10]
        f3=fdist[-10:]
        f2 = len(set(f2) & set(dist_max))
        f3=len(set(f3)&set(dist_min))
        x=[f1,f2,f3]
        user_cmd_feature.append(x)
    return user_cmd_feature

def get_user_cmd_feature_new(user_cmd_list,dist):
    user_cmd_feature=[]

    for cmd_list in user_cmd_list:
        v=[0]*len(dist)
        for i in range(0,len(dist)):
            if dist in cmd_list:
                v[i]+=1
        user_cmd_feature.append(v)

    return user_cmd_feature

def get_label(filename,index=0):
    x=[]
    with open(filename) as f:
        for line in f:
            line=line.strip('\n')
            x.append( int(line.split()[index]))
    return x
```

```python
np.seterr(divide='ignore', invalid='ignore')
user_cmd_list,dist=load_user_cmd_new("../../Data/MasqueradeDat/User3")
user_cmd_feature=get_user_cmd_feature_new(user_cmd_list,dist)
labels=get_label("../../Data/MasqueradeDat/label.txt",2)

y=[0]*50+labels
x_train=user_cmd_feature[0:N]
y_train=y[0:N]

x_test=user_cmd_feature[N:150]
y_test=y[N:150]

# # neigh =KNeighborsClassifier(n_neighbors=3)
# # neigh.fit(x_train, y_train)
# # score = cross_val_score(neigh,x_train+x_test,y_train+y_test,n_jobs=-1,cv=10)
# # # print(neigh,neigh.)
# # print(score)
# # print(np.mean(score))

# # clf=GaussianNB()
# # # clf.fit(x_train,y_train)
# # a = cross_val_score(clf,x_train+x_test,y_train+y_test,n_jobs=-1,cv=10)
# # print(a)
# # print (np.mean(a))

# neigh = KNeighborsClassifier(n_neighbors=5)
# neigh.fit(x_train, y_train)
# print(neigh.score(x_test,y_test)) 

# print("-----------------")
# knn = KNeighborsClassifier(n_neighbors=3)
# neigh.fit(x_train, y_train)

# score = cross_val_score(knn,x_train+x_test,y_train+y_test,cv=5)
# # print(score)
# print(score.mean())

# print("-----------------")


# clf = GaussianNB()
# clf.fit(x_train,y_train)
# y_predict_nb =clf.predict(x_test)
# print(y_predict_nb)


neigh = KNeighborsClassifier(n_neighbors=10)
neigh.fit(x_train, y_train)
y_predict_knn=neigh.predict(x_test)
# print (y_train)

clf = GaussianNB()
clf.fit(x_train, y_train)
y_predict_nb=clf.predict(x_test)
print(y_predict_knn)
print(y_test)
score=np.mean(y_test==y_predict_knn)*100
print ("KNN %d" % score)

print(y_predict_nb)
print(y_test)
score=np.mean(y_test==y_predict_nb)*100
print ("NB %d" % score)
```
